```python
import os
import pandas as pd

# Caminho para o arquivo de texto usando string bruta
file_path = r'C:\Users\Cabas\Desktop\trabalho_ia\pasta\individual_3_3\OE_individual_3_3_0.txt'

# Carregando o arquivo de texto em um DataFrame
df = pd.read_csv(file_path, delim_whitespace=True)

# Verificando as dimensões do DataFrame
print(f'Dimensões do DataFrame: {df.shape}')  # Deve retornar (13, 11)

# Visualizando as primeiras linhas do DataFrame
print(df)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np

s_scaler = StandardScaler()
m_scaler = MinMaxScaler(feature_range=(-1,1))

s_normal_brute_data = s_scaler.fit_transform(df)
m_normal_brute_data = m_scaler.fit_transform(df)

divided_df = df/440

s_normal_divided_data = s_scaler.fit_transform(divided_df)
m_normal_divided_data = m_scaler.fit_transform(divided_df)


```

```python
import matplotlib.pyplot as plt

# Lista de matrizes e títulos
matrices = [df, s_normal_brute_data, m_normal_brute_data, divided_df, s_normal_divided_data, m_normal_divided_data]
titles = ['Brute', 'Std_brute', 'MinMax_brute', 'divided', 'Std_divided', 'MinMax_divided']

# Criar subplots com 2 linhas e 3 colunas
fig, axes = plt.subplots(2, 3, figsize=(12, 10))

# Exibir cada matriz como uma imagem
for i, ax in enumerate(axes.flat):  # axes.flat permite iterar em uma matriz 2D de eixos
    ax.imshow(matrices[i], cmap='gray', interpolation='nearest')
    ax.set_title(titles[i])  # Adicionar título
    ax.axis('off')  # Remover os eixos para deixar a visualização mais limpa

# Ajustar espaçamento entre as subplots
plt.tight_layout()

# Mostrar as imagens
plt.show()
```
